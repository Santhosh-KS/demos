```python
print_points = pd.read_csv("test_lane_points.csv")
#print_points = pd.read_csv("107311171767.csv")
```

```python
print_points['x'] = print_points['x'].astype(int)
print_points['y'] = print_points['y'].astype(int)
```

```python
val = list(zip(print_points.x, print_points.y))
```

```python
source = cv2.imread("test_lane.png")
```

```python
def drawPoints(source, x_y):
    return cv2.circle(source, x_y, int(2), (0,0,255),2, cv2.LINE_AA)
```

```python
for v in val:
    source = drawPoints(source, v)
```

```python
plt.imshow(cv2.cvtColor(source, cv2.COLOR_BGR2RGB))
plt.show()
```
